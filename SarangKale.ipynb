```python
#Que 6) answer is 
class Vehicle:
    
    def __init__(self,model,mileage,price):
        self.price = price
        self.mileage = mileage
        self.model = model
        
    def show_details(self):
        print(f'Model : {self.model}')
        print(f'Price : {self.price}')
        print(f'Mileage : {self.mileage}')
        
        
class Bike(Vehicle):
    def __init__(self,model,mileage,price,tyres,cc):
        super().__init__(model,mileage,price)
        self.cc = cc
        self.tyres = tyres
        
    
    def show_details(self):
        super().show_details()
        print(f'CC : {self.cc}')
        print(f'Tyres : {self.tyres}')
        
```

```python
bajaj = Bike("pulsor",55,225000,2,500)
bajaj.show_details()
```

# Question 3 (15marks Questions) EDA

```python
df = pd.read_csv("supermarketsales.csv")
```

```python
df
```

```python
df.info()
```

```python
(df.isnull().sum()/(len(df)))*100
```

```python
#part1
df.drop(['Invoice ID'],axis=1,inplace=True)
df.drop(['Date'],axis=1,inplace=True)
df.drop(['Time'],axis=1,inplace=True)
df.columns
```

```python
#part2
df.groupby(['Gender','Payment'])['Payment'].count()
```

```python
x= df['Gender']
y= df['Payment']
plt.xlabel("Gender")
plt.ylabel("Payment")
plt.bar(x,y,color='brown',edgecolor='blue',linewidth=1)
plt.show()
```

```python
sns.countplot(x='Gender',hue='Payment',data=df)
plt.show()
```

```python
#part 4
plt.figure(figsize=(12,7))
sns.boxplot(x="Quantity",y="Total",data=df,palette="rainbow")
plt.title("Boxplot of Quantity and Total",size=25)
plt.show()
```

```python
df.describe()
```

```python
#part3
df.sort_values(by='gross income', ascending=False, inplace=True) 
print(df.groupby('Product line').first()) 
```

```python
sns.set_style("whitegrid")
sns.countplot(x="Quantity",hue="gross income",data=df,palette="RdBu_r")
plt.title("Count of Quantity vs gross income",size=25)
plt.show()
```

```python
#part5
sns.heatmap(df.corr(),annot=True,cmap=['orange','green','red','brown'],vmin=-0.5,vmax=0.6)
plt.title("Correlation Table",size=20)
plt.show()
```
